```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Ejemplo de un plot sencillo

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title("Test plot")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

## Ejemplo de dos plot sencillos

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("Test plot")
plt.ylabel("Y and Z")
plt.xlabel("X")
plt.legend(["this is y", "this is z"])
plt.show()
```

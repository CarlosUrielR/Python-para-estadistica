# Generating random samples from $N_p$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Simulating from a $N_2(\boldsymbol{\mu}, \boldsymbol{\Sigma})$

The mean vector and covariance matrix

```python
mean = [0, 0]             # a vector
cov = [[2, -2], [-2, 3]]  # a matrix
```

To check the mean vector and covariance matrix

```python
print(mean)
print(cov)
```

To simulate we can use the function `random.multivariate_normal`

```python
x, y = np.random.multivariate_normal(mean, cov, size=500).T
```

To plot the random sample we can use

```python
plt.plot(x, y, 'x', color='blue')
```

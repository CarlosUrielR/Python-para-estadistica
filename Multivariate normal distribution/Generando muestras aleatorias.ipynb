# Generando muestras aleatorias de una $N_p$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Simulando de una $N_2(\boldsymbol{\mu}, \boldsymbol{\Sigma})$

El vector de medias y la matriz de covarianzas

```python
mean = [0, 0]             # a vector
cov = [[2, -2], [-2, 3]]  # a matrix
```

Para verificar los valores ingresados

```python
print(mean)
print(cov)
```

Para simular se usa la función `random.multivariate_normal`

```python
x, y = np.random.multivariate_normal(mean, cov, size=500).T
```

Para crear el diagrama de dispersión se usa

```python
plt.plot(x, y, 'x', color='blue')
```

Para dibujar una matriz con diagramas de dispersión se usa lo siguiente

```python
d = {'x': x, 'y': y}
df = pd.DataFrame(data=d)
pd.plotting.scatter_matrix(df, color='red')
```

## Simulando de una $N_3(\boldsymbol{\mu}, \boldsymbol{\Sigma})$

```python
mean = [2, -3, 1]             # a vector
cov = [[1, 1, 1], [1, 3, 2], [1, 2, 2]]  # a matrix
```

```python
x, y, z = np.random.multivariate_normal(mean, cov, size=500).T
```

```python
d = {'x': x, 'y': y, 'z': z}
df = pd.DataFrame(data=d)
pd.plotting.scatter_matrix(df, color='orange')
```
